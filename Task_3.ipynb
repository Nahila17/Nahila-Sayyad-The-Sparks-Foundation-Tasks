```python
plt.figure(figsize=(15,10))
sns.countplot(retail['Segment'])
plt.xticks(rotation=90)
plt.show()
```

```python
retail.hist(figsize=(10,10), bins=10)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(20,10)})
sns.lineplot(x="Quantity", y="Profit", data=retail, label='Profit')
plt.show()
```

```python
retail.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['violet','red'],figsize=(10,10))
plt.ylabel('Profit and sales')
plt.show()
```

```python
retail.groupby('State')[['Profit','Sales']].sum().plot.bar(color=['red','purple'],figsize=(10,10))
plt.ylabel('Profit and sales')
plt.show()
```

```python
retail.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['red','purple'],figsize=(10,10))
plt.ylabel('Profit and sales')
plt.show()
```

```python
retail.groupby('Sub-Category')[['Profit','Sales']].sum().plot.bar(color=['red','purple'],figsize=(5,5))
plt.ylabel('Profit and sales')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='Region', y='Sales', data=retail, hue='Segment')
plt.show()
```
